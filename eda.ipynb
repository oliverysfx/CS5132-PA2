```python
df = df[['e10.soil.depth', 'e12.rock.cover', 't01alt', 'awc15', 'bio01', 'bio05', 'bio06', 'gdd', 'nfd',
         'rh410', 't14stot', 'species']]
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
y=df['species'].values
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(df.drop('species', axis=1, inplace=False).values, df['species'].values, test_size=0.2)
```

```python
clf = DecisionTreeClassifier(random_state=0)
```

```python
clf.fit(xtrain, ytrain)
```

```python
from sklearn.metrics import *
```

```python
sns.heatmap(confusion_matrix(ytest, clf.predict(xtest)))
```
